#Gradient backtracking

```python
import numpy as np

def gradient_method_backtracking(f, g, x0, s, alpha, beta, epsilon):
  x = x0
  grad = g(x)
  fun_val = f(x)
  iter = 0
  while  np.linalg.norm(grad) > epsilon:
    iter = iter + 1
    t = s
    while fun_val - f(x - t * grad) < alpha * t * np.linalg.norm(grad)**2:
      t=beta*t
    x = x - t * grad
    fun_val = f(x)
    grad = g(x)
    print('iter_number = {0:3d}'.format(iter),\
          'norm_grad = {0:2.6f}'.format(np.linalg.norm(grad)),\
          'fun_val = {0:2.6f}'.format(fun_val))
  return x, fun_val
```

```python
A = np.array([[1, 0], [0, 0.01]])
f = lambda x : x @ A @ x
g = lambda x : 2 * A @ x
x0 = np.array([2.0, 1.0])
x, fun_val = gradient_method_backtracking(f, g, x0, 2, 0.25, 0.5, 1e-5)
```

```python
ourbees= np.arange(0.001, 1, 0.001).tolist()
```

```python
def gradient_method_backtracking_for_Iter(f, g, x0, s, alpha, beta, epsilon):
  x = x0
  grad = g(x)
  fun_val = f(x)
  iter = 0
  while  np.linalg.norm(grad) > epsilon:
    iter = iter + 1
    t = s
    while fun_val - f(x - t * grad) < alpha * t * np.linalg.norm(grad)**2:
      t=beta*t
    x = x - t * grad
    fun_val = f(x)
    grad = g(x)
  return iter
```

```python
f = lambda x : x @ A @ x
g = lambda x : 2 * A @ x
x0 = np.array([2.0, 1.0])
A = np.array([[1, 0], [0, 0.01]])
itercount=[]
for i in ourbees:
  A = np.array([[1, 0], [0, i]])
  f = lambda x : x @ A @ x
  g = lambda x : 2 * A @ x
  itercount.append(gradient_method_backtracking_for_Iter(f, g, x0, 2, 0.25, 0.5, 1e-5))
```

#Pure Newton

```python
def pure_netwon(f, g, h, x0, epsilon):
  x = x0
  gval = g(x)
  hval = h(x)
  iter = 0
  while np.linalg.norm(gval) > epsilon and iter < 10000:
    iter = iter + 1
    x = x - np.linalg.solve(hval, gval)
    fun_val = f(x)
    gval = g(x)
    hval = h(x)
    print('iter = {0:3d}'.format(iter),\
          'f(x) = {0:2.6f}'.format(fun_val))
    if (iter==10000):
      print('did not converge')
  return x
```

```python
def pure_netwon_for_Iter(f, g, h, x0, epsilon):
  x = x0
  gval = g(x)
  hval = h(x)
  iter = 0
  while np.linalg.norm(gval) > epsilon and iter < 10000:
    iter = iter + 1
    x = x - np.linalg.solve(hval, gval)
    fun_val = f(x)
    gval = g(x)
    hval = h(x)
    if (iter==10000):
      print('did not converge')
  return iter
```
